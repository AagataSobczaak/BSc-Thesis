```python
import numpy as np
import pandas as pd
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
import time
```

# 1 sekundowe pliki

## 13 cech

```python
test_data = pd.read_csv("Sounds_1_seconds/mfcc_1_sec/test_13_mfcc_features_with_labels_1sec.csv")   
y_test = test_data['Encoded_label'] 

unique_labels = y_test.unique()
start_time = time.time()

y_pred_random = np.random.choice(unique_labels, size=len(y_test), replace=True)

end_time = time.time()
classification_time = end_time - start_time

accuracy = accuracy_score(y_test, y_pred_random)
precision = precision_score(y_test, y_pred_random, average='weighted', zero_division=0)
recall = recall_score(y_test, y_pred_random, average='weighted', zero_division=0)
f1 = f1_score(y_test, y_pred_random, average='weighted', zero_division=0)


print("Losowy model bazowy:")
print(f"  Dokładność (Accuracy): {accuracy:.2f}")
print(f"  Precyzja (Precision): {precision:.2f}")
print(f"  Recall: {recall:.2f}")
print(f"  F1 Score: {f1:.2f}")
print(f"  Czas klasyfikacji: {classification_time:.4f} sekund")

print("\nRaport klasyfikacji dla losowego modelu:")
print(classification_report(y_test, y_pred_random, target_names=unique_labels.astype(str)))


sorted_labels = np.sort(unique_labels)
conf_matrix = confusion_matrix(y_test, y_pred_random, labels=sorted_labels)

plt.figure(figsize=(12, 10))
sns.heatmap(conf_matrix, annot=conf_matrix, fmt=".1f", cmap="Greys", 
            xticklabels=sorted_labels, yticklabels=sorted_labels, cbar_kws={'label': 'Liczba przewidywań'})

plt.xlabel("Przewidywanie", fontsize=14)
plt.ylabel("Rzeczywiste", fontsize=14)
plt.title("Macierz pomyłek dla modelu losowego", fontsize=16, fontweight='bold')
plt.xticks(fontsize=12, rotation=45)
plt.yticks(fontsize=12, rotation=0)
plt.tight_layout()
plt.show()
```
